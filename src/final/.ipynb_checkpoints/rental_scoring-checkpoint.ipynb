## Summary

This notebook is a continuation of work in the `rentometer_api.ipynb` and will populate the master properties file `property_level.csv` with MSHDA rent affordablity scores (0 or 5) for user definded rental affordability thresholds of the Rentometer estimated reference apartment rent. The lower the threshold selected, the more likely it is that the calculated rent for a given property (based on number of bedrooms and percentage of county median income) will surpass the Rentometer estimate of a near enough (0.25 mi radius) apartment in Kent County, and thus be considered in need of an affordable housing subsidy and get 5 points for its MSHDA score.

## Create a function to assess rental affordability at different thresholds
This function allows specification of preset affordability bands to account for deviation between actual market realized reant and the Rentometer predicted spot price. As an extreme example, failing to give a property the 5 pro-forma affordbality points because it falls five dollars short of the $ 1,264 1-bedroom hurdle price places far too much confidence in the precision of the Rentometer estimate and fails to capture the reality that rent is always negotiated which makes rent prediction highly subjective.

```python
def rent_affordable(df, br=1, bands=[100]):
    ''' Returns rent affordability assessment (0/1) for preset
        thresholds below the Rentometer API determined rent estimate 
        for apartments in Kent County, MI 
        
    Parameters:
        df: Pandas dataframe with rent estimates (df)
        br: Number of bedrooms (int; 1, 2, 3, 4; must match basis of rent estimate in df)
        bands: List of threshold percentages (int) of rent estimate (default = [100])
        Example: bands = [95, 90, 80, 75]'''
    
    # Copy input dataframe
    df = df.copy()
    
    # Check that br is a valid number
    assert type(br) == int and br > 0 and br < 5, "Valid values of br are: 1, 2, 3, 4"
    
    # Check that br matches basis for #br_rent_estimate column in df
    br_check = [col for col in apartments2_rents if "br_rent_estimate" in col]
    assert br == int(search_col[0][0]), "Please enter value of br that matches basis of rent estimate in df"
    
    # Check if 100 is in bands
    if 100 not in bands:
        bands.append(100)
    
    # Sort bands in ascending order
    bands = sorted(bands)
    
    # Delete all existing affordability columns
    drop_cols = df.columns[df.columns.str.contains(pat = "afford")]
    df = df.drop(columns=drop_cols)
    
    # Create new affordability columns
    for b in bands:
            df[str(br) + "br_affordable_" + str(b)] = None        
    print(df.columns)
    
    # Check affordability threshold at each band
    for i in range(len(df)):
        for b in bands:
            if df[str(br) + "br_rent_estimate"].iloc[i] >= (b/100 * 1264):
                df[str(br) + "br_affordable_" + str(b)].iloc[i] = 1  # this will be converted to 5 points for MSHDA score
            else:
                df[str(br) + "br_affordable_" + str(b)].iloc[i] = 0
    
    # Sum values in each affordability column
    afford_list = []
    
    for b in bands:
        afford_list.append(df[str(br) + "br_affordable_" + str(b)].sum())
            
    return {"bands": bands, "affordable_count": afford_list, "df": df}
```

## Create function to get affordability scores for all properties
Due to insufficient time for debugging, this function could not be used to generate the scores for each rent affordability threshold, so the scores were calculated indiviudally in the code cells below the function

```python
def afforability_banding():
    ''' Returns a dictionary with afforability scores for each property
        based on pre-defined affordabiilty bands'''

    # Join from geom to afford
    afford_geom = geopandas.sjoin(afford,
                                  geom,
                                  how="right",
                                  op="intersects")
    
    afford_geom = afford_geom.reset_index()

    affordable_scores = []

    for b in affordable_bands:
        # Create series and replace all NaN in joined df with 0
        afford_geom_ind = afford_geom.loc[:, b].where(afford_geom.loc[:, b].notna(), 0)

        ID = list(afford_geom.loc[:, "ID"])

        afford_geom = pd.DataFrame({"ID": ID, 
                                    b: afford_geom_ind})
        afford_geom = afford_geom.groupby("ID").agg("sum").reset_index()

        # Convert all affordable values to 0 or 5 and append to list 
        score = afford_geom.loc[:, b].where(afford_geom.loc[:, b].isnull(), 5)
        affordable_scores.append(score)

    # Create dictionary     
    affordable_dict = dict(zip(affordable_bands, affordable_scores))
    
    return affordable_dict
```
